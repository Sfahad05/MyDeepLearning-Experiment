```python
# Import necessary libraries (same as your original)
import pandas as pd
import numpy as np
import re
import os
import torch
import itertools
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords, wordnet as wn
from nltk.stem import WordNetLemmatizer
import nltk
from collections import defaultdict
from sklearn.model_selection import train_test_split, StratifiedKFold
from imblearn.over_sampling import RandomOverSampler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SpatialDropout1D, GRU, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import RMSprop
from sklearn.metrics import (accuracy_score, classification_report,
                             precision_recall_fscore_support, confusion_matrix, roc_curve, auc)
from copy import deepcopy
import matplotlib.pyplot as plt
import seaborn as sns

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî 1. Load & Preprocess Data (identical to your code) ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
try:
    dataset = pd.read_csv('amazonbinaryclassificationdataset_DL.csv', encoding='latin-1')
except Exception as e:
    raise RuntimeError(f"Failed to load dataset: {e}")

print("‚úÖ Dataset loaded.")

# Download NLTK resources
for res in ['punkt', 'wordnet', 'averaged_perceptron_tagger', 'stopwords', 'omw-1.4']:
    nltk.download(res, quiet=True)

# Preprocessing functions (identical)
def strip_html(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

def remove_url(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)

def clean_text(text):
    if not isinstance(text, str):
        return ""
    REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\<\>\|@,;]')
    BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')
    text = strip_html(text)
    text = remove_url(text)
    text = text.lower()
    text = REPLACE_BY_SPACE_RE.sub(' ', text)
    text = BAD_SYMBOLS_RE.sub('', text)
    return text

# Apply preprocessing
dataset['comment_Text'] = dataset['comment_Text'].apply(clean_text)
dataset['comment_Text'].dropna(inplace=True)

# Tokenize & lemmatize (identical)
tag_map = defaultdict(lambda: wn.NOUN)
tag_map['J'] = wn.ADJ
tag_map['V'] = wn.VERB
tag_map['R'] = wn.ADV

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

preprocessed_texts = []
for text in dataset['comment_Text']:
    tokens = word_tokenize(text.lower())
    final_words = []
    for word, pos in pos_tag(tokens):
        if word not in stop_words and word.isalpha():
            lemma = lemmatizer.lemmatize(word, tag_map[pos[0]])
            final_words.append(lemma)
    preprocessed_texts.append(' '.join(final_words))

dataset['text_final'] = preprocessed_texts
dataset['Have_ValueViolation'] = dataset['Have_ValueViolation'].map({'Yes': 1, 'No': 0})
dataset = dataset.dropna(subset=['text_final', 'Have_ValueViolation'])

X_all = dataset['text_final'].values
y_all = dataset['Have_ValueViolation'].astype(int).values

print(f"‚úÖ Preprocessing done. Total samples: {len(X_all)} | Classes: {np.bincount(y_all)}")

# === Device Setup ===
device = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"  # disable GPU for TF
os.environ["TF_ENABLE_ONEDNN_OPTS"] = "0"

# Force TensorFlow to use CPU (critical for stability with PyTorch MPS)
import tensorflow as tf
tf.config.set_visible_devices([], 'GPU')
print("‚úÖ TensorFlow will use CPU only")

# MPS memory workaround
if device.type == "mps":
    os.environ['PYTORCH_MPS_HIGH_WATERMARK_RATIO'] = '0.0'

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî 2. Initial Train/Test Split (Test frozen!) ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
X_temp, X_test, y_temp, y_test = train_test_split(
    X_all, y_all, test_size=0.15, random_state=42, stratify=y_all
)
X_train_raw, y_train_raw = X_temp, y_temp

print(f"\nüìä Splits:")
print(f"‚Üí Train (for CV): {len(X_train_raw)} | Test (held-out): {len(X_test)}")
print(f"‚Üí Train classes (raw): {np.bincount(y_train_raw)} | Test: {np.bincount(y_test)}")

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî 3. GRU Model Definition (identical to your build_GRU_model) ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
def build_gru_model(max_words=1000, embed_dim=100, max_len=150):
    model = Sequential([
        Embedding(input_dim=max_words, output_dim=embed_dim, input_length=max_len),
        SpatialDropout1D(0.2),
        GRU(100, dropout=0.2, recurrent_dropout=0.0),
        Dense(1, activation='sigmoid')
    ])
    model.compile(loss='binary_crossentropy', optimizer=RMSprop(), metrics=['accuracy'])
    return model

# Hyperparameters (same as your script)
max_words = 1000
max_len = 150
batch_size = 100
num_epochs = 50
patience = 5
k_folds = 5

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî 4. K-Fold CV on TRAIN SET (Leakage-Free) ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
print(f"\nüöÄ Starting {k_folds}-Fold CV (on TRAIN only)...")

skf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)
cv_results = []

for fold, (train_idx, val_idx) in enumerate(skf.split(X_train_raw, y_train_raw), 1):
    print(f"\n{'='*50}\n‚ñ∂Ô∏è  FOLD {fold}/{k_folds}\n{'='*50}")

    # Split
    X_tr, X_val = X_train_raw[train_idx], X_train_raw[val_idx]
    y_tr, y_val = y_train_raw[train_idx], y_train_raw[val_idx]

    # ‚úÖ Oversample ONLY fold-train
    ros = RandomOverSampler(random_state=42)
    X_tr_res, y_tr_res = ros.fit_resample(X_tr.reshape(-1, 1), y_tr)
    X_tr_res = X_tr_res.flatten().tolist()
    y_tr_res = y_tr_res.astype(int)

    print(f"Fold {fold} ‚Üí Train: {len(X_tr_res)} (oversampled), Val: {len(X_val)}")
    print(f"  ‚Üí Train classes: {np.bincount(y_tr_res)} | Val: {np.bincount(y_val)}")

    # ‚úÖ Tokenizer FIT ONLY on oversampled fold-train
    tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
    tokenizer.fit_on_texts(X_tr_res)

    def prepare_seqs(texts):
        seqs = tokenizer.texts_to_sequences(texts)
        return pad_sequences(seqs, maxlen=max_len, padding='post', truncating='post')

    X_tr_seq = prepare_seqs(X_tr_res)
    X_val_seq = prepare_seqs(X_val)

    # ‚úÖ Fresh GRU model per fold
    model = build_gru_model(max_words, 100, max_len)
    early_stop = EarlyStopping(
        monitor='val_accuracy',
        patience=patience,
        restore_best_weights=True,
        verbose=0
    )

    # Train
    history = model.fit(
        X_tr_seq, y_tr_res,
        batch_size=batch_size,
        epochs=num_epochs,
        validation_data=(X_val_seq, y_val),
        callbacks=[early_stop],
        verbose=0
    )

    # Evaluate on original val fold
    y_val_pred = (model.predict(X_val_seq, verbose=0).flatten() > 0.5).astype(int)
    val_acc = accuracy_score(y_val, y_val_pred)
    precision, recall, f1, _ = precision_recall_fscore_support(
        y_val, y_val_pred, average='binary', zero_division=0
    )

    cv_results.append({
        'fold': fold,
        'epochs_trained': len(history.history['loss']),
        'val_accuracy': val_acc,
        'precision': precision,
        'recall': recall,
        'f1': f1
    })

    print(f"‚úÖ Fold {fold} | Epochs: {len(history.history['loss'])} | "
          f"Val Acc: {val_acc:.4f} | F1: {f1:.4f}")

# Summarize CV
cv_df = pd.DataFrame(cv_results)
print("\n" + "="*70)
print("üìä K-FOLD CV RESULTS (mean ¬± std):")
print(f"Accuracy : {cv_df['val_accuracy'].mean():.4f} ¬± {cv_df['val_accuracy'].std():.4f}")
print(f"Precision: {cv_df['precision'].mean():.4f} ¬± {cv_df['precision'].std():.4f}")
print(f"Recall   : {cv_df['recall'].mean():.4f} ¬± {cv_df['recall'].std():.4f}")
print(f"F1-Score : {cv_df['f1'].mean():.4f} ¬± {cv_df['f1'].std():.4f}")
print("="*70)

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî 5. Final Model: Retrain on FULL TRAIN ‚Üí Test on HELD-OUT TEST ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
print("\nüéØ Retraining final GRU on FULL TRAIN (oversampled)...")

# Oversample full training set
ros_full = RandomOverSampler(random_state=42)
X_train_res, y_train_res = ros_full.fit_resample(X_train_raw.reshape(-1, 1), y_train_raw)
X_train_res = X_train_res.flatten().tolist()
y_train_res = y_train_res.astype(int)

# Tokenizer fit on full oversampled train
tokenizer_final = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer_final.fit_on_texts(X_train_res)

def prepare_seqs_final(texts):
    seqs = tokenizer_final.texts_to_sequences(texts)
    return pad_sequences(seqs, maxlen=max_len, padding='post', truncating='post')

X_train_seq = prepare_seqs_final(X_train_res)
X_test_seq = prepare_seqs_final(X_test)

# Final model
final_model = build_gru_model(max_words, 100, max_len)

# Use small internal val for early stopping (from train only)
X_tr_final, X_val_final, y_tr_final, y_val_final = train_test_split(
    X_train_seq, y_train_res, test_size=0.1, random_state=42, stratify=y_train_res
)

early_stop_final = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

history_final = final_model.fit(
    X_tr_final, y_tr_final,
    batch_size=batch_size,
    epochs=num_epochs,
    validation_data=(X_val_final, y_val_final),
    callbacks=[early_stop_final],
    verbose=1
)

# ‚úÖ FINAL TEST EVAL (held-out, natural distribution)
y_test_pred_proba = final_model.predict(X_test_seq, verbose=0).flatten()
y_test_pred = (y_test_pred_proba > 0.5).astype(int)

test_acc = accuracy_score(y_test, y_test_pred)
precision, recall, f1, _ = precision_recall_fscore_support(
    y_test, y_test_pred, average='binary'
)

print(f"\nüü© FINAL TEST RESULTS (Held-Out, Natural Distribution):")
print(f"Overall Accuracy          : {test_acc:.4f}")
print(f"Violation Precision       : {precision:.4f}")
print(f"Violation Recall          : {recall:.4f}")
print(f"Violation F1-Score        : {f1:.4f}")

print("\nüìã Classification Report (Test Set ‚Äî 4 decimals):")
print(classification_report(y_test, y_test_pred,
                            target_names=["No Violation", "Violation"],
                            digits=4))

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî 6. Plotting (identical to your style, enhanced labels & DPI) ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
def plot_history(hist):
    acc = hist.history['accuracy']
    val_acc = hist.history['val_accuracy']
    loss = hist.history['loss']
    val_loss = hist.history['val_loss']
    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(14, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, 'b-o', label='Train Accuracy')
    plt.plot(epochs, val_acc, 'r-o', label='Validation Accuracy')
    plt.title('GRU: Training & Validation Accuracy')
    plt.xlabel('Epochs'); plt.ylabel('Accuracy'); plt.legend(); plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, 'b-o', label='Train Loss')
    plt.plot(epochs, val_loss, 'r-o', label='Validation Loss')
    plt.title('GRU: Training & Validation Loss')
    plt.xlabel('Epochs'); plt.ylabel('Loss'); plt.legend(); plt.grid(True)
    plt.tight_layout()
    plt.savefig('gru_learning_curves.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_confusion_matrix(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=classes, yticklabels=classes)
    plt.title("GRU: Confusion Matrix (Final Test Set)")
    plt.ylabel('True Label'); plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig('gru_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_roc_curve(y_true, y_pred_proba):
    fpr, tpr, _ = roc_curve(y_true, y_pred_proba)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC (AUC = {roc_auc:.3f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0]); plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
    plt.title('GRU: ROC Curve (Final Test Set)')
    plt.legend(loc="lower right"); plt.grid(True)
    plt.tight_layout()
    plt.savefig('gru_roc_curve.png', dpi=300, bbox_inches='tight')
    plt.show()

# Generate all plots
plot_history(history_final)
plot_confusion_matrix(y_test, y_test_pred, ["No Violation", "Violation"])
plot_roc_curve(y_test, y_test_pred_proba)
```
